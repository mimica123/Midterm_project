```python
import pandas as pd
import numpy as np
```

```python
#importing data
df=pd.read_csv('C:/Users/bd/lighthouse-data-notes/Week_5/Day_5/mid-term-project-I/flights_clean1.csv')
pd.set_option('display.max_columns',100)
df.head()
```

```python
#Picking features for our model
data=df[['mkt_unique_carrier','op_unique_carrier','mkt_carrier_fl_num','origin_airport_id','dest_airport_id','dep_time','crs_dep_time','crs_arr_time','crs_elapsed_time','dep_delay','taxi_out','wheels_off','wheels_on','taxi_in','arr_time','actual_elapsed_time','air_time','distance','year','month','minutes','hours','arr_delay']]
data.head()
```

# Creating new features
## origin

```python
origin=pd.DataFrame(data[['origin_airport_id','arr_delay']].groupby('origin_airport_id').sum().reset_index())
origin.sort_values(by='arr_delay',ascending=False)
```

```python
#Calculating the probability of delay due to the origin
origin['proba_delay']=origin['arr_delay']/len(data)*100
```

```python
weight=[]
for i in data['origin_airport_id']:
    weight.append(float(origin[origin['origin_airport_id']==i]['proba_delay']))
```

```python
# add it to the data
data['proba_delay_origin']=weight
```

```python
data.head()
```

## Destination

```python
destination=pd.DataFrame(data[['dest_airport_id','arr_delay']].groupby('dest_airport_id').sum().reset_index())
destination.sort_values(by='arr_delay',ascending=False)
```

```python
#Calculating the probability of delay due to the destination
destination['proba_delay']=destination['arr_delay']/len(data)*100
```

```python
#Now we are going to calculate the weight
weight=[]
for i in data['dest_airport_id']:
    weight.append(float(destination[destination['dest_airport_id']==i]['proba_delay']))
```

```python
# add it to the data
data['proba_delay_dest']=weight
```
