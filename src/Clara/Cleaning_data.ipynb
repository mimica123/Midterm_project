```python
import pandas as pd
import numpy as np
```

```python
#importing data
df=pd.read_csv('C:/Users/bd/lighthouse-data-notes/Week_5/Day_5/flights.txt')
pd.set_option('display.max_columns',100)
df.head()
```

```python
#verifying null elements in the features
df.isnull().sum()
```

```python
#creating a copy of the data
data=df.copy()
```

```python
#For null in delays it will be filled by 0
data['arr_delay'].fillna(0)
data['dep_delay'].fillna(0)
```

```python
#getting year and month from fl_date column
from datetime import datetime
data['fl_date'] = pd.to_datetime(data['fl_date'],format='%Y-%m-%d')
data['year'] = pd.DatetimeIndex(data['fl_date']).year
data['month'] = pd.DatetimeIndex(data['fl_date']).month
```

```python
#crs_arr_time doesn't have missing values and arr_time has
#we are going to inpute the same time that crs_arr_time in arr_time
#for the missing values

import math

i=0
while i <10000:
    if math.isnan(data['arr_time'][i]):
        data['arr_time'].loc[i]=data['crs_arr_time'][i]
    i+=1
data['arr_time'].isnull().any()
```

```python
#crs_dep_time doesn't have missing values and dep_time has
#we are going to inpute the same time that crs_dep_time in dep_time
#for the missing values

i=0
while i <10000:
    if math.isnan(data['dep_time'][i]):
        data['dep_time'].loc[i]=data['crs_dep_time'][i]
    i+=1
data['dep_time'].isnull().any()
```

```python
#crs_elapsed_time doesn't have missing values and actual_elapsed_time has
#we are going to inpute the same time that crs_elapsed_time in actual_elapsed_time
#for the missing values

i=0
while i <10000:
    if math.isnan(data['actual_elapsed_time'][i]):
        data['actual_elapsed_time'].loc[i]=data['crs_elapsed_time'][i]
    i+=1
data['actual_elapsed_time'].isnull().any()
```

```python
#getting hour and minutes from dep_time and arr_time
data['minutes']=data['dep_time']%100

trunc = lambda x: math.trunc(x / 100);
data['hours']=data['dep_time'].apply(trunc)

data['minutes']=data['arr_time']%100

trunc = lambda x: math.trunc(x / 100);
data['hours']=data['arr_time'].apply(trunc)
```

```python
#droping rows wit nan in columns air_time and arr_delay
data.dropna(subset = ["air_time",'arr_delay','dep_delay'], inplace=True)
data.isnull().sum()
```

```python
#change to a category to convert to numbers
#for mkt_unique_carrier
data["mkt_unique_carrier"] = data["mkt_unique_carrier"].astype('category')
data["mkt_unique_carrier"] = data["mkt_unique_carrier"].cat.codes

#for mkt_unique_carrier
data["op_unique_carrier"] = data["op_unique_carrier"].astype('category')
data["op_unique_carrier"] = data["op_unique_carrier"].cat.codes
```

```python
#looking for normalization in column arr_delay
from scipy.stats import shapiro
from scipy.stats import normaltest

_,n = normaltest(data['arr_delay'])
_,s = shapiro(data['arr_delay'])
print('Normal(p)=%.3f, shapire(p)=%.3f' % (n, s))
```

```python
data.to_csv('flights_clean1.csv')
```
