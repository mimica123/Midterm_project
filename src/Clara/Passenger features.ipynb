```python
#importing data
df=pd.read_csv('C:/Users/bd/lighthouse-data-notes/Week_5/Day_5/mid-term-project-I/passengers.txt')
pd.set_option('display.max_columns',100)
df.head()
```

```python
data=df.copy()
```

# Passanger per month

```python
passenger=data[['month','passengers']].groupby('month').sum().reset_index()
passenger.sort_values('passengers', ascending=False)
```

```python
#Calculating the percentage of passengers per month
passenger['perc_passenger']=passenger['passengers']/passenger['passengers'].sum()*100
weight=[]
for i in data['month']:
    weight.append(float(passenger[passenger['month']==i]['perc_passenger']))
```

```python
# add it to the data
data['perc_passenger']=weight
```

```python
data.head()
```

# Passenger per carrier

```python
carrier=data[['unique_carrier','passengers']].groupby('unique_carrier').sum().reset_index()
carrier.sort_values('passengers', ascending=False)
```

```python
#Calculating the percentage of passengers per carrier
carrier['perc_passenger_carrier']=carrier['passengers']/carrier['passengers'].sum()*100
weight=[]
for i in data['unique_carrier']:
    weight.append(float(carrier[carrier['unique_carrier']==i]['perc_passenger_carrier']))
```

```python
# add it to the data
data['perc_passenger_carrier']=weight
```
