```python
def missing_values(column,replace):
    #################################
    # Column has missing values, we are going to change those values
    # using replace
    # Expected value:
    #      -A dataframe with missing values
    #      -A datafame with values we want to replace
    # Return:
    #      -A dataframe without missing values
    #      
    #
    # Example:  missing_values([10,0.2, nan, 5.3],[10.3,0.2,3,5.1])
    #
    # Return: [10,0.2, 3, 5.3]
    #################################
    import pandas as pd
    import numpy as np
    import math

    i=0
    while i < len(column):
        if math.isnan(column[i]):
            column.loc[i]=replace[i]
        i+=1
    return column
```

```python
def get_hour(time):
    #################################
    # From a column that contain hour and minutes together in the
    # the same row in the format %H%M
    # Expected value:
    #      -A dataframe with time
    #     
    # Return:
    #      -A dataframe with hour
    #      -A dataframe with minutes
    #
    # Example:  get_hour(1708)
    #
    # Return: 17, 08
    #################################
    import math
    minutes=time%100
    print(time%100)
    trunc = lambda x: math.trunc(x / 100);
    hours=time.apply(trunc)
    return hours, minutes
```

```python
def categorize(data):
    #################################
    # Changing a feature to a catagory
    # Expected value:
    #      -A dataframe 
    #     
    # Return:
    #      -A dataframe categorized
    #
    # Example:  categorize([10,12,5,1,7]) in 3 categories
    #
    # Return: [2,2,1,0,1]
    #################################
    data = data.astype('category')
    data = data.cat.codes
    return data
```
