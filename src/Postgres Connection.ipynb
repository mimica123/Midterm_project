```python
# Importing modules to use
import psycopg2 as pg
import pandas as pd
import json
```

```python
# Load the pg connection from config.json
with open('../config.json') as f:
    cfg = json.loads(f.read())
```

```python
# Connect to postgres
pg_cfg = cfg['postgres']
conn = pg.connect(**pg_cfg)
cur = conn.cursor()
```

```python
# Get data from database
sql = """SELECT * FROM flights LIMIT 100"""
cur.execute(sql)
flights = cur.fetchall()
# Get column names
df = pd.DataFrame(flights, columns=[c.name for c in cur.description])
df
```

```python
# Write results to csv file
df.to_csv('../data/100flights.csv', index=False)
```

```python
# Clean up
cur.close()
conn.close()
```
